## Machine learning 

##### Use decision tree

##### Based on the observations, how many more people are likely to use laboratory services in the future 

```python
survey1.info()
```

##### drop some unnecessary columns

```python
survey1=survey1.drop(['1. What is your age range?','2. What is your Gender?','3. Where is your current location? (city, country)','5. Who referred you to the Laboratory?', '7. Please describe your experience briefly?','11. How did you resolve it?','12. How likely are you to continue using the Laboratory services in the future.','19. If you could see something improved about laboratory services, what would it be?'], axis='columns')
survey1
```

### rename columns

```python
survey1.columns=['visit','process','rate','staff','mix_up','recommend','book','preference','collection','satisfication']
survey1.head(3)
```

### Converting some columns from words to digit using labelEncoder and fit transform

```python
from sklearn.preprocessing import LabelEncoder

le_visit=LabelEncoder()
le_process=LabelEncoder()
le_mix_up=LabelEncoder()
le_recommend=LabelEncoder()
le_book=LabelEncoder()
le_preference=LabelEncoder()
le_collection=LabelEncoder()
le_satisfication=LabelEncoder()
```

```python
survey1['visit']=le_visit.fit_transform(survey1['visit'])
survey1['process']=le_process.fit_transform(survey1['process'])
survey1['mix_up']=le_mix_up.fit_transform(survey1['mix_up'])
survey1['recommend']=le_recommend.fit_transform(survey1['recommend'])
survey1['book']=le_book.fit_transform(survey1['book'])
survey1['preference']=le_preference.fit_transform(survey1['preference'])
survey1['collection']=le_collection.fit_transform(survey1['collection'])
survey1['satisfication']=le_satisfication.fit_transform(survey1['satisfication'])
```

```python
survey1.head(5)
```

## Dropping out the target variable

```python
input=survey1.drop(['satisfication'],axis='columns')
target=survey1['satisfication']
target
```

```python
input.head(3)
```

importing the model

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input,target)
```

training the model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.3)
```

```python
model.score(input,target)
```
